```python
import easyocr           #for recognition of character
import cv2               
import numpy as np
import glob             #return all file paths that match a specific pattern
```

```python
path = '/home/neosoft/Documents/ProjectCV/CarImages/*.*'      #folder path
```

```python
reader = easyocr.Reader(['en'], gpu=False)
output = []

for file in glob.glob(path):
    img = cv2.imread(file, 0)       #0 means reads the image as a gray image
    result = reader.readtext(img, paragraph=False)
    text = ""
    
    for i in range(len(result)):
        if len(result[i][1])>7 and len(result[i][1])<14:
            print(result[i][1])
            text += result[i][1]
            break
            
    if len(text)<1:
        output.append((file, "Not detected"))
    else:
        output.append((file, text))
        
output   
```

```python
img2 = 'car10.jpg'
 
#read the image
image1 = cv2.imread('car4.jpeg')

#dividing height and width by 2 to get the center of the image
height, width = image1.shape[:2]

#get the center coordinates of the image to create the 2D rotation matrix
center = (width/2, height/2)

#using cv2.getRotationMatrix2D() to get the rotation matrix
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=180, scale=1)

#rotate the image using cv2.warpAffine
rotated_image = cv2.warpAffine(src=image1, M=rotate_matrix, dsize=(width, height))

#visualize the original and the rotated image
cv2.imshow('Original image', image1)
cv2.imshow('Rotated image', rotated_image)

#wait indefinitely, press any key on keyboard to exit
cv2.waitKey(0)

#write the output, the rotated image to disk
status = cv2.imwrite('/home/neosoft/Documents/ProjectCV/RotatedImages/rotated_image2.jpg', rotated_image)
print("Image written to file-system : ",status)
```
